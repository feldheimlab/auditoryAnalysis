```python
import numpy as np
import scipy.io
import sys
# import stimgen
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
```

# Stimulation file


```python
stimpath = '/home/feldheimlab/Documents/AuditoryStimGeneration/cfos_behavior_stim_threechannel.wav'
rate, data = wavfile.read(stimpath)
print(rate)
N = int(rate/2)
T = 1.0 / rate
x = np.linspace(0.0, N*T, N, endpoint=False)
time = x/rate


yf = fft(data[:N, 0])
xf = fftfreq(N, T)[:N//2]/1000
yfnoise = fft(data[-N:, 0])
xfnoise = fftfreq(N, T)[:N//2]/1000

x = np.linspace(0.0, N*T, N, endpoint=False)
time = x/rate

dB = 20*np.log10(20*np.log10(np.max(data[:N])/np.max(data[-N:])))
rms_calc = 1/np.sqrt(2)
print(rms_calc)
print(np.max(data[:N, 0]))
print((np.max(data[-N:,0])))
print(20*np.log10(np.max(data[:N, 0])/(np.max(data[-N:,0])*rms_calc)))
```

## FFT

```python
fig, axs = plt.subplots(2,1,figsize=(10,5))

axs[0].plot(time[:N], data[:N, 0], color='black')
axs[0].plot(time[:N], data[-N:, 0], color='red')
axs[0].set_ylim(-2**16/2,2**16/2)

smoothy = np.convolve(2.0/N * np.abs(yf[0:N//2]), np.ones(100), mode='same')
smoothynoise = np.convolve(2.0/N * np.abs(yfnoise[0:N//2]), np.ones(100), mode='same')

axs[1].plot(xfnoise, smoothynoise, color='red', alpha=1, label='background')
axs[1].plot(xf,smoothy, color='black', alpha=1, label='stimulation')

axs[1].set_xticks([5,20,40,60,80,100])
axs[1].set_xlim(0,110)
axs[1].set_ylim(10**0,10**5)
axs[1].set_yscale('log')
    
axs[0].set_ylabel('intensity')
axs[1].set_ylabel('power')
axs[1].set_xlabel('frequency (kHz)')
axs[1].legend()
plt.savefig('stimulation_file.png')
plt.show()
```

```python
binary = np.arange(0,(2**16)/2)
voltages = binary * 10000 * 2/(2**16) #converted to 16 bit to voltage (10,000mV/32768 binary; set by NI code)  
print(voltages[-1])
ref = 80 #80mV == 1Pa
dBV = 20*np.log10(voltages/ref)
db = dBV + 94 - 40 #94db=1Pa, +40db amplifier from mic 

fig  = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()
ax3 = ax1.twinx()

ax1.plot(voltages, db, color='b')#voltage to dB SPL
ax1.set_xlabel('voltages (mV)')
ax1.set_ylabel('dB (SBL)')
ax1.grid('on')

ax2.plot(binary, db, color='b') #top axis in binary values
ax2.set_xlabel('binary value')

ax3.plot(voltages, dBV, color='b') #right y axis in dBV
ax3.set_ylabel('dBV')

plt.savefig('dB_conversions.png', dpi=300)
plt.show()
```

## FFT

```python
from scipy.fft import fft, fftfreq

import numpy as np


# Number of sample points
N = int(rates[0]/2)
# sample spacing

T = 1.0 / 500000.0
x = np.linspace(0.0, N*T, N, endpoint=False)
conv = (10000 * 2)/(2**16)

fig = plt.figure(figsize=(5,5))

labels = ['high', 'low']
db_dataset = np.zeros((2,2,3))

fig, axs = plt.subplots(2,2,figsize=(10,5))
a = -1
ref = 80
# conversion  = 98ref/10 #1
for d, data in enumerate(datas):
    if d % 3 == 0:
        a += 1
        axs[0][a].plot(time[:N], data[:N]*conv, color='black', alpha=1)
        axs[0][a].plot(time[:N], data[-N:]*conv, color='red', alpha=1)
#         axs[0][a].set_ylim(-1500*conv,1500*conv)

    yf = fft(data[:N] * conv)
    xf = fftfreq(N, T)[:N//2]/1000
    yfnoise = fft(data[-N:] * conv)
    xfnoise = fftfreq(N, T)[:N//2]/1000

    smoothy = np.convolve(2.0/N * np.abs(yf[0:N//2]), np.ones(100), mode='same')
    smoothynoise = np.convolve(2.0/N * np.abs(yfnoise[0:N//2]), np.ones(100), mode='same')
        
    smoothy = 20*np.log10(smoothy/ref) #+ 94 - 40
    smoothynoise = 20*np.log10(smoothynoise/ref) #+ 94 - 40
    
    if d % 3 == 0:
        axs[1][a].plot(xfnoise, smoothynoise, color='red', alpha=0.25, label='background')
        axs[1][a].plot(xf,smoothy, color='black', alpha=0.25, label='stimulation')
    else:
        axs[1][a].plot(xfnoise, smoothynoise, color='red', alpha=0.25)
        axs[1][a].plot(xf,smoothy, color='black', alpha=0.25)
    axs[1][a].set_xticks([5,20,40,60,80,100])
    axs[1][a].set_xlim(0,100)

    axs[1][a].title.set_text(labels[a])
    
    rms = np.max(data[:N]*conv)*0.707
    db_dataset[0,a,d%3] = 20*np.log10(rms/ref) #+ 94 - 40
    rms = np.max(data[-N:]*conv)*0.707
    db_dataset[1,a,d%3] = 20*np.log10(rms/ref) #+ 94 - 40
    
axs[0][0].set_ylabel('mV')
axs[1][0].set_ylabel('dB')
axs[1][1].set_xlabel('frequency (kHz)')
axs[1][1].legend()
plt.savefig('TDT-speaker_recording_stimulation_file.png')
plt.show()
dbmean = np.mean(db_dataset, axis=2)
dbstd = np.std(db_dataset, axis=2)
for m, mean in enumerate(dbmean[0]):
    print('Signal is {0} +/- {1} dB'.format(np.round(mean,2), np.round(dbstd[0,m], 2)))
print('\n')
for m, mean in enumerate(dbmean[1]):
    print('Noise is {0} +/- {1} dB'.format(np.round(mean,2), np.round(dbstd[1,m], 2)))
```
